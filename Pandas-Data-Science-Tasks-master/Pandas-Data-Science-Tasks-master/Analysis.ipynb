### ADD MONTH COLUMN

```python
all_data['Month']=all_data['Order Date'].str[0:2]
```

```python
all_data[all_data.isna().any(axis=1)]
```

```python
all_data=all_data.dropna(how='all')
```

```python
all_data[all_data.isna().any(axis=1)]
```

```python
all_data=all_data[all_data['Order Date'].str[0:2]!='Or']
```

```python
all_data['Month']=all_data['Month'].astype('int32')
```

### CREATING A COLUMN SALES

```python
all_data.head()
```

#### WE NEED TO CONVERT STRING FROM TO INT FORM

```python
all_data['Quantity Ordered']=all_data['Quantity Ordered'].astype('int32')
all_data['Price Each']=pd.to_numeric(all_data['Price Each'])
```

```python
all_data['Sales']=all_data['Quantity Ordered']*all_data['Price Each']
```

### Which is the best month for sales?

```python
result = all_data.groupby('Month').sum('Sales')
```

### PLOTING THE RESULTS ON A BAR GRAPH

```python
import matplotlib.pyplot as plt
months=range(1,13)
plt.bar(months,result['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD')
plt.xlabel('Months')
plt.show()
```

### CITY WITH HIGHEST NO: OF SALES

```python
all_data.head()
```

```python
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(" ")[1]
all_data['City']=all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
```

### INSERTING A NEW COLUMN NAMED CITY 

```python
all_data.head()
```

```python
result=all_data.groupby('City').sum('Sales')
result.drop('Month',axis=1,inplace=True)
```

```python
result.head()
```

```python
import matplotlib.pyplot as plt
cities=[city for city,df in all_data.groupby('City')]
plt.bar(cities,result['Sales'])
plt.xlabel('City')
plt.ylabel('Sales')
plt.xticks(cities,rotation='vertical')
plt.show()
```
